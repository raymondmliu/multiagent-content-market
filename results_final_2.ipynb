```python
import matplotlib.pyplot as plt
import numpy as np

from model_classes import InfoAccessEnum
from run_model import run_model, graph_model, plot_heatmap, plot_supps

functions = {
    'descending rapidly': (lambda x: np.exp(- 4 * x)),
    'linear': (lambda x: 1 - 0.5 * x),
    'descending slowly': (lambda x: 1 + np.exp(-8) - np.exp(4 * x - 8)),
}
```

```python
for f in functions:
    model = run_model(
        num_members=9,
        M_INFL=5.0,
        M=5.0,
        f=functions[f],
    )

    plot_heatmap(model, f"consumer function {f}", f"f/{f}")
```

```python
for f in functions:
    model = run_model(
        num_members=9,
        M_INFL=10.0,
        M=10.0,
        g=functions[f],
    )

    plot_heatmap(model, f"producer function {f}", f"g/{f}")
```

```python
for m in [4.0, 5.0, 6.0]:
    model = run_model(
        num_members=9,
        M_INFL=m,
        f=functions['descending slowly'],
        M=5.0,
    )

    plot_heatmap(model, f"influencer rate budget = {m}", f"M_INFL/{m}")
```

```python
for t in [InfoAccessEnum.LIMITED, InfoAccessEnum.PERFECT]:
    model = run_model(
        num_members=9,
        M_INFL=5.0,
        M=5.0,
        info_access=t,
        f=functions['descending slowly'],
    )

    plot_heatmap(model, f"info access {InfoAccessEnum(t).name}", f"lim/{t}")
```
